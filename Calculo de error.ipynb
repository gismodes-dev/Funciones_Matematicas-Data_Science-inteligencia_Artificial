```python
import matplotlib.pyplot as plt
```

```python
x=[1.2, 2,3.2,2.5,5,6,4,8]
y=[2,3,3.4,3.1,4,4.7,3.8,7]
y_prima=[]
```

##Calculo de la regresion Lineal

Sabiendo que:
$$f(x)=mx + b$$ donde $m$ y $b$  $\in R$. 

entonces:

image.png

```python
def sum_xy(x,y):
  xy=0
  for i in range(len(x)):
      xy+= x[i]*y[i]
  return xy
```

```python
def sum_x2(x):
  x2=0
  for i in range(len(x)):
      x2+= x[i]**2
  return x2
```

## prendiente *m*


```python
def calc_m(x,y):
  numerador=len(x)*sum_xy(x,y)-(sum(x)*sum(y))
  denominador=len(x)*sum_x2(x)-(sum(x))**2
  return numerador/denominador
```

## interseccion *b*

```python
def calc_b(x,y):
  m=calc_m(x,y)
  y_prom=sum(y)/len(y)
  x_prom=sum(x)/len(x)
  return y_prom-(m*x_prom)
```

## Funcion lineal

```python
def f(x,m,b):
  return m*x+b
```

## Resultado de los valores de y'

```python
m=calc_m(x,y)
b=calc_b(x,y)

for x_val in x:
  y_prima.append(round(f(x_val,m,b),2))

print(y_prima)
```

# Grafico 

```python
#recta para graficar
x_fig=[x for x in range(0,11)]
y_fig=[]
for val in x_fig:
  y_fig.append(round(f(val,m,b),2))
```

```python

fig,ax=plt.subplots()
ax.plot(x_fig,y_fig, label='y\'=mx+b')
# ax.scatter(x,y_prima,color='g',label='y\'')
ax.scatter(x,y,color='r', label="y")
ax.set_ylim(0,8)
ax.set_xlim(0,9)
ax.grid()
ax.legend(loc ="lower right")
fig.show()
```

# Error cuadratico medio ECM

```python
print(y)
print(y_prima)
```

```python
def calc_ecm(y,y_prima):
  sum=0
  for i in range(len(y)):
    sum+=(y_prima[i]-y[i])**2
  return sum/len(y)
```

```python
emc=calc_ecm(y,y_prima)
print(f'el error cuadratico medio es {emc}' )
```
